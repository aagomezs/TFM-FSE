```python
# Importar Librerias
import pandas as pd
import os
```

```python
# 📁 Ruta donde están tus archivos
ruta_carpeta = "G:\Mi unidad\TFM UNIR\BBDD TFM"
```

```python
# 📄 Lista de archivos
archivos = [f for f in os.listdir(ruta_carpeta) if f.endswith(".csv")]

# 🔄 Lista donde guardaremos los dataframes
lista_df = []

for archivo in archivos:
    ruta_completa = os.path.join(ruta_carpeta, archivo)
    
    print(f"Leyendo archivo: {archivo}")
    
    # df = pd.read_csv(ruta_completa, sep=";", encoding="latin1")
    df = pd.read_csv(ruta_completa, sep=";")
    
    # 🧠 Opcional: agregar columna con nombre del archivo (útil para trazabilidad)
    df["origen_archivo"] = archivo
    
    lista_df.append(df)

# 🔗 Unir todos los dataframes
df_consolidado = pd.concat(lista_df, ignore_index=True)

# 💾 Guardar resultado
df_consolidado.to_csv("dataset_consolidado.csv", index=False)

print("✅ Consolidación finalizada. Archivo generado: dataset_consolidado.csv")
```
